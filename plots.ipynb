```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_numworkers = pd.read_csv("mpi_measurements_numworkers.csv")
data_size = pd.read_csv("mpi_measurements_size.csv")
```

```python
plt.figure()
print("Зависимость времени работы от числа воркеров")
x = list(data_numworkers['numworkers'])
y = list(data_numworkers['time'])
plt.plot(x, y, 'bo-')
plt.show()
```
